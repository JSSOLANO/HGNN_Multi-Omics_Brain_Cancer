```python
import os
import glob
import pandas as pd
```

```python
#OBTAIN THE LIST OF FILE IN THE NUCLEOTIDE FOLDER
path = '/Users/sebastiansolano/Documents/AI_projects/HANN_Oncogene/TCGA_project_data_for_training/CPTAC-3_Brain-nos_sinlge_nucleotide_variation/'
file_pattern = os.path.join(path, '*')
all_files = glob.glob(file_pattern)
```

```python
#CREATE A LIST WITH AL DATASET
df_list = (pd.read_csv(file, sep='\t') for file in all_files)
```

```python
#CONCATENATE THE DATASETS
combined_df = pd.concat(df_list, ignore_index=True)
```

```python
#LIST THE CATEGORIES IN VARIAN CALSSIFICATION
variant_category=combined_df["Variant_Classification"].unique()
variant_category=variant_category.tolist()
```

```python
#ELIMMINATE THE NOT NECCESARY VARIANT CATEGORIES
remove_list=["3'Flank","5'Flank",'RNA','Silent']
for remove in remove_list:
    variant_category.remove(remove)
```

```python
#FILTER THE DATABSE TO WORK ONLY WITH THE "MAIN" VARIANT CATEGORIES 
Variant_dataset_trainig=combined_df[combined_df["Variant_Classification"].isin(variant_category)]
```

```python
#VERIFY NO SILENT MUTATION WERE INCLUDE
Variant_dataset_trainig[Variant_dataset_trainig['Variant_Classification']=='Silent']
```
